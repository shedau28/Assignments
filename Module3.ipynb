 Write a Python function to check whether a number is perfect or not.


```python
def perfect_number(n):
    sum = 0
    for x in range(1, n):
        if n % x == 0:
            sum += x
    return sum == n
print(perfect_number(5))
```

Write a Python function that checks whether a passed string is palindrome 
or not

```python
def fun2(num):
  str1 = str(num)
  str2 = ''
  l1 = []
  for i in str1:
    l1.append(i)
  l1.reverse()
  for j in l1:
    str2 += j
  if str1 == str2:
    print('Pelindrome')
  else:
    print("not Pelindrome")


fun2(1234)
```

Write a Python program to returns sum of all divisors of a number

```python
def sum_div(number):
    divisors = [1]
    for i in range(2, number):
        if (number % i)==0:
            divisors.append(i)
    return sum(divisors)
print(sum_div(8))
print(sum_div(12))
```
